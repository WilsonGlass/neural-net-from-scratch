```python
"""
Randomly changing and searching for optimal weights and biases did not prove fruitful.
The number of possible combinations fo weights and biases is infinite,a nd we need
something smarter than luck. Each weight and bias may also have different degrees of
influence on the loss, this influence depends on the parameters themselves as well as on
the current sample, which is an input to the first layer.

Here we will describe exactly how gradient descent works. We will be calculating much
each singular weight and bias changes the loss value.

Our goal is to decrease loss, and we'll do this by using gradient descent. Gracient, on the
other hand, is a result of the calculation of the partial derivatives, and we'll backpropogate
it using the chain rule to update all the weights and biases.
"""

import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2 * x ** 2

x = np.array(range(5))
y = f(x)
print(x)
print(y)
plt.plot(x, y)
plt.show()

# TODO: This chapter and the next seem kind of simple, I think I will skim these and move onto
# chapter 9 so I'm not reteaching myself calculus.
```
